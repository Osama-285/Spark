```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Accumulator").getOrCreate()
spark
```

```python
df = spark.read.csv("Flight_Schedule.csv", inferSchema = True, header = True)
df.show(5)
df.printSchema()
```

```python
sc = spark.sparkContext
rdd = df.rdd
nullCount = sc.accumulator(0)

def nullCountFunction(row):
    if row["scheduledArrivalTime"] is None:
        nullCount.add(1)
    else:
        return

rdd.foreach(nullCountFunction)
print(f"Total Null Arrival Times: {nullCount.value}")
print(f"Total Null Arrival Times: {nullCount}")
```

```python
spark.stop()
```
